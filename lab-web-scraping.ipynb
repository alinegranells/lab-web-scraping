```python
def get_book_details(book_url):
    
    response = requests.get(book_url)
    soup = BeautifulSoup(response.text, "html.parser")

    # UPC
    upc = soup.find("th", string="UPC").find_next("td").text

    # Genre
    genre = soup.find("ul", class_="breadcrumb").find_all("a")[2].text

    # Availability
    availability = soup.find(
        "p",
        class_="instock availability"
    ).text.strip()

    # Description
    description_header = soup.find(
        "div",
        id="product_description"
    )

    if description_header:
        description = description_header.find_next("p").text
    else:
        description = None

    return {
        "UPC": upc,
        "Genre": genre,
        "Availability": availability,
        "Description": description
    }
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

base_url = "https://books.toscrape.com/catalogue/page-{}.html"

headers = {
    "User-Agent": "Mozilla/5.0"
}

rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

all_books = []

for page in range(1, 51):

    print(f"Scraping page {page}")

    url = base_url.format(page)

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
    except Exception as e:
        print(f"Error on page {page}: {e}")
        continue

    soup = BeautifulSoup(response.text, "html.parser")

    books = soup.find_all("article", class_="product_pod")

    for book in books:

        # ----- BASIC INFO -----
        title = book.h3.a["title"]

        price_text = book.find("p", class_="price_color").text
        price = float("".join(c for c in price_text if c.isdigit() or c == "."))

        rating_class = book.find("p", class_="star-rating")["class"][1]
        rating = rating_map[rating_class]

        relative_url = book.h3.a["href"]
        book_url = "https://books.toscrape.com/catalogue/" + relative_url

        # ----- DETAIL PAGE -----
        try:
            book_response = requests.get(book_url, headers=headers, timeout=10)
            book_soup = BeautifulSoup(book_response.text, "html.parser")

            table = book_soup.find("table")

            product_info = {}
            for row in table.find_all("tr"):
                key = row.find("th").text
                value = row.find("td").text
                product_info[key] = value

            upc = product_info.get("UPC")

            genre = book_soup.find("ul", class_="breadcrumb").find_all("a")[2].text

            availability = book_soup.find("p", class_="instock availability").text.strip()

            description_tag = book_soup.find("div", id="product_description")
            description = (
                description_tag.find_next("p").text
                if description_tag else None
            )

            time.sleep(0.2)

        except Exception as e:
            print(f"Error on book: {title} -> {e}")
            continue

        # ----- STORE -----
        record = {
            "UPC": upc,
            "Title": title,
            "Price (£)": price,
            "Rating": rating,
            "Genre": genre,
            "Availability": availability,
            "Description": description
        }

        all_books.append(record)

        time.sleep(0.1)

# ----- FINAL DATAFRAME -----
df = pd.DataFrame(all_books)

df.head()
```
